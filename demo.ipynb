```python
# Gerekli modülleri içe aktar
import sys, os
sys.path.insert(0, os.path.abspath('.'))

from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 120
%matplotlib inline

from core.task     import Task, make_task
from core.scorer   import compute_score, rank_tasks, score_breakdown, print_ranking
from core.scheduler import schedule, schedule_random, schedule_fcfs
from core.evaluator import evaluate, evaluate_all_scenarios
from data.synthetic import sample_tasks, load_scenario
from viz.charts    import gantt_chart, comparison_bar, score_distribution

NOW = datetime(2025, 1, 1, 9, 0)   # sabit zaman → tekrar üretilebilir
print('✅ Modüller yüklendi.')
```

## 1. Örnek Görev Listesi ve Skor Hesaplama

```python
tasks = sample_tasks(now=NOW)

print(f'Toplam görev: {len(tasks)}\n')
print_ranking(tasks)
```

```python
# Tek görev için skor bileşenleri
print(score_breakdown(tasks[0]))
```

## 2. Skor Dağılımı

```python
fig = score_distribution(tasks, title='Örnek Görevler — Skor Dağılımı')
plt.show()
```

## 3. AI Sistemi — Gantt Chart

```python
result_ai = schedule(tasks, available_hours=8.0, start_time=NOW)

print(result_ai.summary())
print()

fig = gantt_chart(result_ai, title='Örnek Görevler')
plt.show()
```

## 4. Senaryo Karşılaştırması
Aynı görev seti **Random**, **FCFS** ve **AI (Greedy+EDF)** ile planlanır.

```python
for scenario_name in ('rahat', 'yogun', 'imkansiz'):
    tasks_s = load_scenario(scenario_name, now=NOW)
    report  = evaluate(tasks_s, scenario_name=scenario_name,
                       available_hours=8.0, start_time=NOW)
    report.print_table()
```

## 5. Görsel Karşılaştırma Grafikleri

```python
for scenario_name in ('rahat', 'yogun', 'imkansiz'):
    tasks_s = load_scenario(scenario_name, now=NOW)
    report  = evaluate(tasks_s, scenario_name=scenario_name,
                       available_hours=8.0, start_time=NOW)
    fig = comparison_bar(report)
    plt.show()
```

## 6. Yoğun Senaryo — Gantt Chart Karşılaştırması

```python
tasks_yogun = load_scenario('yogun', now=NOW)

r_ai  = schedule       (tasks_yogun, available_hours=8.0, start_time=NOW)
r_rnd = schedule_random(tasks_yogun, available_hours=8.0, start_time=NOW)
r_fcs = schedule_fcfs  (tasks_yogun, available_hours=8.0, start_time=NOW)

for result in (r_ai, r_rnd, r_fcs):
    fig = gantt_chart(result, title='Yoğun Senaryo')
    plt.show()
```

## 7. Sonuç

| Algoritma | Güçlü Yön | Zayıf Yön |
|-----------|-----------|----------|
| **AI (Greedy+EDF)** | Yüksek önemli görevleri öne alır, deadline'lara uyar | Deterministik; aynı input → aynı output |
| **Random** | - | Önemli görevler gecikebilir |
| **FCFS** | Basit ve adil | Önem/aciliyet gözetmez |

**Sonuç cümlesi:**  
> *Sistem, NP-hard scheduling problemine karşılık olarak heuristic + greedy + EDF hibrit yaklaşımı ile pratik ve ölçeklenebilir bir çözüm sunmaktadır.*
